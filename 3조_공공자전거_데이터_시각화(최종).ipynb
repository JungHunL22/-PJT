# 날씨 데이터 전처리

```python
## 읽기
weather_data = pd.read_csv('./data/OBS_ASOS_TIM_20220818094831.csv',encoding='cp949')
```

```python
## 제거
weather_data = weather_data.drop(['지점','지점명'], axis=1)
```

```python
weather_data
```

```python
# 대여일자, 시간 분할
date=[]
time=[]
for i in weather_data.일시:
    date.append(i.split(' ')[0])
    time.append(i.split(' ')[1].split(':')[0])
weather_data['대여일자']=date
weather_data['대여시간']=time
```

```python
weather_data
```

```python
# 순서 변환
weather_data = weather_data[['대여일자','대여시간','기온(°C)','강수량(mm)','적설(cm)','풍속(m/s)','일시']]
```

```python
# 일시 열 제거
weather_data = weather_data.drop(['일시'], axis=1)
```

```python
weather_data
```

```python
# 우천 여부
weather_data['우천여부']=['no' if i==0 else 'yes' for i in weather_data['강수량(mm)']]
weather_data
```

```python
# 시간대
weather_data['시간대']=['출/퇴근' if i=='07' or i=='08' or i=='09' or i=='17' or i=='18' or i=='19' else '출/퇴근 외' for i in weather_data['대여시간']]
weather_data
```

```python
# 계절
month=[i.split('-')[1] for i in weather_data['대여일자']]
season=[]
for i in month:
    if i=='03' or i=='04' or i=='05':
        season.append('봄')
    elif i=='06' or i=='07' or i=='08':
        season.append('여름')
    
    elif i=='09' or i=='10' or i=='11':
        season.append('가을')
    else:
        season.append('겨울')
weather_data['계절']=season
```

```python
weather_data
```

```python
# datetime 변환
weather_data['대여일자']=pd.to_datetime(weather_data['대여일자'])
```

```python
## 휴일
holidays_in_2020 = pytimekr.holidays(year=2020)
holidays_in_2020
def return_holidays(x):
    if x in holidays_in_2020:
        return '휴일'
    elif x.weekday() in range(5):
        return '평일'
    elif x.weekday() in range(5,8):
        return '휴일'

weather_data['평일/휴일']=weather_data['대여일자'].apply(return_holidays)
```

```python
weather_data
```

```python
bicycle_total_used
```

```python
bicycle_total_used.reset_index(inplace=True)
# bicycle_total_used.to_csv('./data/일별자전거이용건수.csv',encoding='cp949', index=False)
```

```python
# 이용건수 병합
weather_data['대여일자'] = weather_data['대여일자'].astype(str)
weather_data['대여시간'] = weather_data['대여시간'].astype(int)
weather_data = pd.merge(weather_data, bicycle_total_used, how='left')
```

```python
weather_data
```

```python
## 시간당 이용률을 위해 일별 합계 계산
ration=pd.pivot_table(weather_data,index='대여일자',values='이용건수',aggfunc='sum').reset_index()
ration['일 합계']=ration.이용건수
ration.drop('이용건수',axis=1,inplace=True)
ration
```

```python
weather_data = pd.merge(weather_data,ration,how='left',on='대여일자')
```

```python
weather_data
```

```python
weather_data['시간당 이용률']=(weather_data.이용건수/weather_data['일 합계'])*100
weather_data
```

```python
weather_data = weather_data.drop(['일 합계'], axis=1)
```

```python
# 결측 처리
weather_data['기온(°C)'].fillna(33.7,inplace=True)
weather_data=weather_data.fillna(0)
weather_data.isnull().sum()
```

## 대여소 별 이용건수 상위 100개 데이터 시각화

```python
# 상위 100개 데이터 시각화
latitude=37.5502 # 첫 위치 지정
longitude=126.982
import folium
from folium.plugins import MarkerCluster

m = folium.Map( # 지도 그리기
    location=[latitude, longitude],
    zoom_start=10.5
)

coords = ranks[['위도','경도']] # 상위 100개 위도 경도 값 저장

marker_cluster = MarkerCluster().add_to(m) # 마커

for lat, long in zip(coords['위도'], coords['경도']): # 표시
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
# 구별 이용건수
gu_map=ranks[['자치구','이용건수']]
gu=gu_map.groupby('자치구')['이용건수'].sum()
print(gu)
```

```python
import json

geo_path='./data/02. skorea_municipalities_geo_simple.json'
geo_str=json.load(open(geo_path,encoding='utf-8')) # json 불러오기


bins = list(gu.quantile([0, 0.25, 0.5, 0.75, 1])) # 백분위수 list
m = folium.Map( # 지도 그리기
    location=[37.559819, 126.963895],
    zoom_start=11, 
    tiles='cartodbpositron'
)

folium.GeoJson( # 위치 지정
    geo_str,
    name='지역구'
).add_to(m)

m.choropleth(geo_data=geo_str, # 시각화
             data=gu, 
             fill_color='YlOrRd',
             fill_opacity=0.5,
             line_opacity=0.2,
             key_on='properties.name',
             legend_name='구별 TOP100 대여소 이용건수'
            )
m
```
